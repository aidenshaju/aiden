```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# 1. Read the dataset to the python environment.
# The Iris dataset is available directly within scikit-learn
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target] # Add species names

display(df.head())
```

```python
# 2. Display the columns in the dataset.
print("\nColumns in the dataset:")
print(df.columns)
```

```python
# 3. Calculate the mean of each column of the dataset.
print("\nMean of each numerical column:")
print(df.drop('species', axis=1).mean()) # Exclude the 'species' column for mean calculation
```

```python
# 4. Check for the null values present in the dataset.
print("\nNull values in the dataset:")
print(df.isnull().sum())
```

```python
# 5. Perform meaningful visualizations using the dataset.
# Visualization 1: Pairplot to show relationships between features and species
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue='species', diag_kind='kde')
plt.suptitle('Pairplot of Iris Dataset Features by Species', y=1.02)
plt.show()
```

```python
# Visualization 2: Box plots for each feature by species
plt.figure(figsize=(12, 6))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(1, 4, i + 1)
    sns.boxplot(y=feature, x='species', data=df)
    plt.title(f'Box Plot of {feature}')
plt.tight_layout()
plt.show()
```

```python
# Visualization 3: Scatter plot of sepal length vs. sepal width colored by species
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=df, s=100)
plt.title('Sepal Length vs. Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)
plt.show()
```
